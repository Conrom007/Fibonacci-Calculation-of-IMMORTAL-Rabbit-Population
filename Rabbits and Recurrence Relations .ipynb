```python
content=open('/content/rosalind_fib.txt')
f=content.read().split(' ')
n=int(f[0])
k=int(f[1])
content.close()

def rabbits(n,k):
    prev1 = 1
    prev2 = 1
    for i in range(2, n):
        current = prev1 + k * prev2
        prev2 = prev1
        prev1 = current
    return current

rabbits(n,k)


```
